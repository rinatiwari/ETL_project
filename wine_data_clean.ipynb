```python
# Dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
#import and read second csv file
csv_file = "data/winemag-data-v2.csv"
wine_review_df = pd.read_csv(csv_file,  encoding = 'iso-8859-1', low_memory=False)
wine_review_df.head()
```

```python
#create new datafame with new colum name in second csv file
wine_review_df.rename(columns={'Unnamed: 0':'id'}, inplace=True)
wine_review_df.head()
```

```python
#create new data with selected columns
new_wine_review_df = wine_review_df[['id','country','points','price', 'province', 'region_1','region_2','title','variety', 'winery']].copy()
new_wine_review_df.head()
```

```python
# Create a list of the columns
columns = ['id','country','points','price', 'province', 'region_1','region_2','title','variety']
    

#  Create a new df for "US" with the columns. 
wine_review_us =new_wine_review_df.loc[new_wine_review_df["country"] == "US", columns]
wine_review_us.head()
```

```python
 #  Create a new df for "France" with the columns. 
wine_review_france =new_wine_review_df.loc[new_wine_review_df["country"] == "France", columns]
wine_review_france
```

```python
#  Create a new df for "Italy" with the columns. 
wine_review_italy =new_wine_review_df.loc[new_wine_review_df["country"] == "Italy", columns]
wine_review_italy
```

```python
#merge dataframes fo "us" and "france"
merged_wine_review_df=pd.merge(wine_review_us, wine_review_france, how="outer", on="id")
merged_wine_review_df
```

```python
#connect to local database
rds_connection_string = "postgres:postgres@localhost:5432/Wine_DB"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
#check table
engine.table_names()
```

```python
#load dataframe "us review" into database
wine_review_us.to_sql(name='wine_review_us', con=engine, if_exists='append', index=False)
```

```python
#load dataframe "france review" into database
wine_review_france.to_sql(name='wine_review_france', con=engine, if_exists='append', index=False)
```

```python
#Confirm data has been added by querying the tables
pd.read_sql_query('select * from wine_review_us', con=engine).head()
```

```python
#Confirm data has been added by querying the tables
pd.read_sql_query('select * from wine_review_france', con=engine).head()
```
