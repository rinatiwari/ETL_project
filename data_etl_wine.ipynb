```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "winemag-data.csv"
wine_review_df = pd.read_csv(csv_file)
wine_review_df.head()
```

```python
wine_review_df.rename(columns={'Unnamed: 0':'id'}, inplace=True)
wine_review_df.head()
```

```python
csv_file = "winemag-datav2.csv"
wine_review_v2_df = pd.read_csv(csv_file)
wine_review_v2_df.head()
```

```python
wine_review_v2_df.rename(columns={'Unnamed: 0':'id'}, inplace=True)
wine_review_v2_df.head()
```

```python
wine_review = wine_review_df[['id', 'country', 'points', 'price', 'province', 'region_1', 'region_2', 'variety', 'winery']].copy()
wine_review.head()
```

```python
wine_review_v2 = wine_review_v2_df[['id', 'country', 'points', 'price', 'province', 'region_1', 'region_2', 'title', 'variety', 'winery']].copy()
wine_review_v2.head()
```

```python
wcf_connection_string = "postgres:postgres@localhost:5432/Wine_DB"
engine = create_engine(f'postgresql://{wcf_connection_string}')
```

```python
engine.table_names()
```

```python
wine_review.to_sql(name='wine_reviews', con=engine, if_exists='append', index=False)
```

```python
wine_review_v2.to_sql(name='wine_reviews_v2', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from wine_reviews', con=engine).head()
```

```python
pd.read_sql_query('select * from wine_reviews_v2', con=engine).head()
```
